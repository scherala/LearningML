### Chi2 Statistic

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.chi2.html#sklearn.feature_selection.chi2

from sklearn.feature_selection import chi2
import pandas as pd # For reading data
import warnings
warnings.filterwarnings(action="ignore")

# Read data
heart_data = pd.read_csv("https://raw.githubusercontent.com/scherala/LearningML/master/Data/heart.csv")

# Separate features
nominal_features = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']
target_column = ['DEATH_EVENT']
numeric_features = set(heart_data.columns) - set(nominal_features) - set(target_column)

# Get Chi2 statistic
chi2_statistic, pvalues = chi2(heart_data[nominal_features], heart_data[target_column])

print("Chi2 Statistic: " )
print(chi2_statistic)

print("P Values:")
print(pvalues)
```

### Covariance and Correlation Coefficient:

```python
heart_data_numeric = heart_data[numeric_features].join(heart_data[target_column])
heart_data_numeric.cov()
```

```python
heart_data_numeric.corr()
```
